# **Exploratory Data Analysis**

##**Gender**

```python
ax = sns.countplot(x = 'Gender', data = df, palette = 'Set2', edgecolor = 'k', linewidth = 1.3 )

for col in ax.containers:
  ax.bar_label(col)
plt.xlabel( 'Gender', style = 'italic', fontsize = 8, )
plt.ylabel('Count', style = 'italic', fontsize = 8,  )
plt.xlabel('Gender', style = 'italic', weight = 'bold', fontsize = 10)
plt.title('Gender Distribution', style = 'italic', weight = 'bold', fontsize = 15, color = 'violet')
plt.show()
```

```python
df.groupby(['Gender'], as_index = False )['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

From the above bargraph we can clearly see the lot of purchesers are
females and even the purchasing power greater than men

## **Age**

```python
plt.figure(figsize = (7, 7))
ax = sns.countplot(x = 'Age Group', data = df, palette = 'Set1', edgecolor = 'k')

for col in ax.containers:
  ax.bar_label(col)

plt.xlabel('Age Group', style = 'italic', fontsize = 10)
plt.ylabel('Count', style = 'italic', fontsize = 10)
plt.title('Age Group Distribution', fontsize = 12, weight = 'bold')

plt.show()
```

```python
df.groupby('Age Group' , as_index = False )['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

From above there is 26-35 age group have significantly high purchasing power then 36-45 and then 1-25 and so on...


### **Age and Gender**

```python
c = [ 'pink', 'blue']

ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender', palette = c, edgecolor = 'k', linewidth = 1.3)
plt.xlabel('Age Group', style = 'italic',  fontsize = 10)
plt.ylabel('Count', style = 'italic',  fontsize = 10)
plt.title('Age Group Distribution', fontsize = 12, weight = 'bold')
legend = plt.legend(title='Gender', )

for col in ax.containers:
  ax.bar_label(col)
# Get the text objects (labels) from the legend
legend_labels = legend.get_texts()
legend_title = legend.get_title()
legend_title.set_color('darkgreen')

# Set the color for each label individually
legend_labels[0].set_color('pink') # Assuming 'M' is the first label
legend_labels[1].set_color('blue')  # Assuming 'F' is the second label

plt.tight_layout()
plt.show()
```

```python
df.groupby(['Age Group', 'Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

from the above its clear females have more purchasing power in age wise too.

## **State**

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(15)

sns.set(rc = {'figure.figsize': (15, 5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders', palette = 'Set2')
plt.xticks(rotation = 40)
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(15)

sns.set(rc = {'figure.figsize': (15, 5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount', palette = 'Set2')
plt.xticks(rotation = 40)
plt.show()
```

## **Marital Status**

```python
plt.figure(figsize = (3,6))
ax = sns.countplot(data = df, x = 'Marital_Status', palette = 'Set1', edgecolor = 'k', linewidth = 1.3)

for col in ax.containers:
  ax.bar_label(col)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize': (4, 5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue = 'Gender', palette = 'Set2', edgecolor = 'k')
plt.xticks(np.arange(2), ('F', 'M'))
legend_label = legend.get_texts()

legend_labels[0].set_color('pink')
legend_labels[1].set_color('blue')
plt.show()
```

    Woman has bought the most and have greater purchasing power

###**Occupation**

```python
plt.figure(figsize = (15, 5))
ax = sns.countplot(data = df, x = 'Occupation', palette = 'Set1', edgecolor = 'k', linewidth = 1.3)

for col in ax.containers:
  ax.bar_label(col)

plt.xticks(rotation = 30)
plt.show()
```

From the above table we can conclude that most buyer and purchaser are in IT,
HealthCare, Aviation and Banking.

### **Product Category**

```python
df.head()
```

```python
plt.figure(figsize= (15, 5))

sns.countplot(data = df, x = 'Product_Category', palette = 'viridis', edgecolor = 'k', linewidth = 1.3)

for col in ax.containers:
  ax.bar_label(col)

plt.xticks(rotation = 90)

plt.show()
```

```python
product_sales = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

plt.figure(figsize = (15, 5))
sns.barplot(data = product_sales, x = 'Product_Category', y = 'Amount', palette = 'Set2', edgecolor = 'violet')
plt.xticks(rotation = 90)
plt.show()
```

From the above table the most sold products are food, clothing and electronics

```python
sales_state = df.groupby('Product_ID', as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)

plt.figure(figsize = (10, 5))
sns.barplot(data = sales_state.head(10), x = 'Product_ID', y = 'Orders', palette = 'Set2', edgecolor = 'k')
plt.xticks(rotation = 90)

plt.show()
```

```python
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values( ascending = False).plot(kind = 'bar', figsize= ( 10, 5), color = 'blue')
```
